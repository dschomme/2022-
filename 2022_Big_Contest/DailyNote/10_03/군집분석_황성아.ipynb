```python
from kmodes.kprototypes import KPrototypes
import plotly.express as px
```

```python
target = merge['product_id']
data = merge.drop(columns=['application_id','user_id','loan_rate','loan_limit','gender','yearly_income','insert_time','income_type','employment_type','houseown_type','purpose','desired_amount','personal_rehabilitation_yn','personal_rehabilitation_complete_yn',	'existing_loan_cnt','existing_loan_amt','reage','career'])
```

```python
data
```

```python
cost = []
for cluster in range(3,7):
    print('{cluster}\.cluster start'.format(cluster=cluster))
    kproto = KPrototypes(n_clusters = cluster, init='Huang',n_init=10,verbose= True)
    kproto.fit(data, categorical= [0, 1])
    cost.append(kproto.cost_)
```

```python
kproto = KPrototypes(n_clusters = 5, init='Huang',n_jobs=-1,n_init=10,verbose=True)
kproto.fit(data, categorical=[0,1])
centroids = kproto.cluster_centroids_
labels  = kproto.labels_
```
