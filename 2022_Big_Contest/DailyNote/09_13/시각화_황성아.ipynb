# Loan_result

```python
loan_result = pd.read_csv('/content/gdrive/MyDrive/BigContest_data/loan_result.csv')
loan_result
```

# Loan_result, User_spec 합치기 (outer)

```python
merge1 = pd.merge(loan_result, user_spec,
                    how = 'outer',
                    on = 'application_id')
merge1
```

```python
# 데이터 정보 확인
print(merge1.info())
print(merge1.isnull().sum())
```

```python
merge1.describe()
```

```python
merge1['is_applied'].unique()
```

```python
f, ax = plt.subplots(figsize=(7, 7))
merge1['is_applied'].value_counts().plot.pie(ax= ax, autopct='%1.1f%%',textprops={'fontsize': 12} )
```

-> 타깃 데이터 불균형

```python
plt.figure(figsize=(9,9))
corr = merge1.corr()
sns.heatmap(corr, cmap='RdBu')
```
