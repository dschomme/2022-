```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
final = pd.read_csv('/Users/sungahwang/Desktop/BC_data/final.csv')
final.head()
```

```python
a = final[final['cluster2'] == 0.0]
b = final[final['cluster2'] == 1.0]
c = final[final['cluster2'] == 2.0]
d = final[final['cluster2'] == 3.0]
e = final[final['cluster2'] == 4.0]
f = final[final['cluster2'] == 5.0]
```

```python
fig, ax = plt.subplots(ncols=3, nrows=2)
sns.set(rc={'figure.figsize':(30,30)})

a['is_applied'].value_counts().plot.pie(ax= ax[0,0], autopct='%1.1f%%',textprops={'fontsize': 20} )
b['is_applied'].value_counts().plot.pie(ax= ax[0,1], autopct='%1.1f%%',textprops={'fontsize': 20 } )
c['is_applied'].value_counts().plot.pie(ax= ax[0,2], autopct='%1.1f%%',textprops={'fontsize': 20 } )
d['is_applied'].value_counts().plot.pie(ax= ax[1,0], autopct='%1.1f%%',textprops={'fontsize': 20} )
e['is_applied'].value_counts().plot.pie(ax= ax[1,1], autopct='%1.1f%%',textprops={'fontsize': 20 } )
f['is_applied'].value_counts().plot.pie(ax= ax[1,2], autopct='%1.1f%%',textprops={'fontsize': 20 } )
```
