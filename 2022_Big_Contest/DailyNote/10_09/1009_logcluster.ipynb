```python
from sklearn.cluster import KMeans
```

```python
def elbow(X):
  sse = []
  for i in range(1,11):  
    km = KMeans(n_clusters=i, init='k-means++', random_state = 0)
    km.fit(X)
    sse.append(km.inertia_)
  
  plt.plot(range(1,11), sse, marker ='o')
  plt.xlabel('클러스터 개수')
  plt.ylabel('SSE')
  plt.show()

elbow(log_data)
```

```python
model = KMeans(n_clusters=6)
model.fit(log_data)
```

```python
print (model.inertia_)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_transformed = pca.fit_transform(log_data)

log_data['pca_x'] = pca_transformed[:,0]
log_data['pca_y'] = pca_transformed[:,1]
log_data.head(3)
```

```python
log_data['cluster']=model.labels_
```

```python
marker0_ind = log_data[log_data['cluster']==0].index
marker1_ind = log_data[log_data['cluster']==1].index
marker2_ind = log_data[log_data['cluster']==2].index
marker3_ind = log_data[log_data['cluster']==3].index
marker4_ind = log_data[log_data['cluster']==4].index
marker5_ind = log_data[log_data['cluster']==5].index

# cluster값 0, 1, 2에 해당하는 Index로 각 cluster 레벨의 pca_x, pca_y 값 추출. o, s, ^ 로 marker 표시
plt.scatter(x=log_data.loc[marker0_ind,'pca_x'], y=log_data.loc[marker0_ind,'pca_y'], marker='o') 
plt.scatter(x=log_data.loc[marker1_ind,'pca_x'], y=log_data.loc[marker1_ind,'pca_y'], marker='s')
plt.scatter(x=log_data.loc[marker2_ind,'pca_x'], y=log_data.loc[marker2_ind,'pca_y'], marker='^')
plt.scatter(x=log_data.loc[marker3_ind,'pca_x'], y=log_data.loc[marker3_ind,'pca_y'], marker='*')
plt.scatter(x=log_data.loc[marker4_ind,'pca_x'], y=log_data.loc[marker4_ind,'pca_y'], marker='+')
plt.scatter(x=log_data.loc[marker5_ind,'pca_x'], y=log_data.loc[marker5_ind,'pca_y'], marker='<')


plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('6 Clusters Visualization by 2 PCA Components')
plt.show()
```
