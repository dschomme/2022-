```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# 데이터 불러오기
loan_result = pd.read_csv('/Users/sungahwang/Desktop/BigContest_data/loan_result.csv')
loan_result.head(3)
```

```python
loan_result.isnull().sum()
```

```python
loan_target = loan_result[loan_result.is_applied.isnull()]
loan_target
```

```python
loan_target.isnull().sum()
```

```python
loan_target = loan_target.dropna(subset = ['loan_limit', 'loan_rate'])
loan_target
```

```python
loan_target.isnull().sum()
```

```python
loan_train = loan_result[(loan_result['is_applied']==0)|(loan_result['is_applied']==1)]
loan_train
```

```python
loan_train.isnull().sum()
```

-> train에 있는 결측값은 채워야함 !!

```python
loan_result = pd.concat([loan_target, loan_train])
loan_result
```

## 파생변수 weekday 추가

```python
loan_result['loanapply_insert_time'] = pd.to_datetime(loan_result['loanapply_insert_time'])
```

```python
loan_result['weekday'] = loan_result['loanapply_insert_time'].dt.day_name()
loan_result.head(3)
```
